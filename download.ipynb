```python
# DOWNLOAD VARIABLES

# low quality (to show possibilities):
download_all = False
show_all = True

# high quality (dowload specific):
download_index = -1     # -1 -> no download

# info
path_to_image_by_index = 4
```

```python
from sentinelsat import SentinelAPI, make_path_filter
from datetime import date

api = SentinelAPI(None, None)

def getQuery(down_left, top_right, start_date, end_date):
    footprint = "POLYGON(("
    footprint += down_left[0] + " " + down_left[1] + ","
    footprint += down_left[0] + " " + top_right[1] + ","
    footprint += top_right[0] + " " + top_right[1] + ","
    footprint += top_right[0] + " " + down_left[1] + ","
    footprint += down_left[0] + " " + down_left[1]
    footprint += "))"

    return api.query(footprint,
                    date=(start_date, end_date),
                    platformname='Sentinel-2',
                    processinglevel='Level-2A',
                    cloudcoverpercentage=(0, 20))

# Mazury
products = getQuery(("20.860291", "53.633239"), ("22.118225", "54.059388"), date(2021, 7, 1), date(2021, 7, 31))

products_df = api.to_dataframe(products)
products_df
```

```python
def download_all_images():
    # download only B02, B03, B04 from 60m/px
    path_filter = make_path_filter("*B0[234]_60m.jp2", exclude = False)
    api.download_all(products, directory_path='downloads', nodefilter = path_filter)

if (download_all):
    download_all_images()
```

```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np


def get_path(path_idx, scale='10'):
    path = 'downloads/'
    path += products_df.to_dict('split')['data'][path_idx][0] 
    path += '.SAFE/GRANULE/L2A_'
    path += products_df.to_dict('split')['data'][path_idx][0].split('_')[5]
    path += '_'
    path += products_df.to_dict('split')['data'][path_idx][22].split('_')[7]
    path += '_'
    path += products_df.to_dict('split')['data'][path_idx][38].split('_')[7][1:]
    path += '/IMG_DATA/R' + scale + 'm/'
    path += products_df.to_dict('split')['data'][path_idx][0].split('_')[5]
    path += '_'
    path += products_df.to_dict('split')['data'][path_idx][0].split('_')[2]
    return path
    

def load_sentinel_image(img_folder, filename, bands, scale):
    image = {}
    for band in bands:
        file = img_folder + 'R' + scale + 'm/' + filename + '_' + band + '_' + scale + 'm.jp2'
        ds = rasterio.open(file)
        image.update({band: ds.read(1)})

    return image


def display_rgb(img, b_r, b_g, b_b, alpha = 1., figsize = (10, 10), title=''):
    rgb = np.stack([img[b_r], img[b_g], img[b_b]], axis=-1)
    rgb = rgb/rgb.max() * alpha
    plt.figure(figsize = figsize)
    plt.title(title)
    plt.imshow(rgb)

def show_all_images():
    for i in range(len(products_df.to_dict('split')['data'])):
        try:
            filepath = get_path(i, '60')
            img = load_sentinel_image(filepath[:filepath.index("IMG_DATA") + 9], filepath[filepath.index("IMG_DATA") + 14:], ['B02', 'B03', 'B04'], '60')
            display_rgb(img, 'B04', 'B03', 'B02', alpha = 5., figsize = (4, 4), title = str(i))
            print(str(i) + ' good')
        except:
            print('error: ' + str(i))

if (show_all):
    show_all_images()
```

```python
# download one image by id:
def download_by_id(id):
    path_filter = make_path_filter('*B??_??m.jp2', exclude = False)
    api.download(products_df.to_dict('split')['data'][id][41], directory_path='downloads', nodefilter = path_filter)

    print('Your PATH to image:')
    print(get_path(id))

    filepath = get_path(id, '60')
    img = load_sentinel_image(filepath[:filepath.index("IMG_DATA") + 9], filepath[filepath.index("IMG_DATA") + 14:], ['B02', 'B03', 'B04'], '60')
    display_rgb(img, 'B04', 'B03', 'B02', alpha = 5., figsize = (4, 4), title = str(id))

if (download_index != -1):
    download_by_id(download_index)

if (path_to_image_by_index != -1):
    print()
    print('path to image (index = ' + str(path_to_image_by_index) + '):')
    print(get_path(path_to_image_by_index))
```
